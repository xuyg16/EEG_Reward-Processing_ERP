# Make Feedback Epochs
run feedback-locked epochs, and save them

```python
from pathlib import Path
import sys

cwd = Path.cwd().resolve()
REPO_ROOT = next(path for path in (cwd, *cwd.parents) if (path / 'scripts' / 'config.py').exists())
SCRIPTS_DIR = REPO_ROOT / 'scripts'
if str(SCRIPTS_DIR) not in sys.path:
    sys.path.insert(0, str(SCRIPTS_DIR))

import config
EPOCHS_ROOT = REPO_ROOT / 'output_mne' / 'epochs'

from decoding.decoding_utils.epoch_io import _load_bids_raw, build_and_save_feedback_epochs, get_epochs_path
import pandas as pd
```

```python
PIPELINE = 'proposed'
USER = 'xu'
SUBJECTS = sorted(config.SUBJECT_INFO)
ONLY_MISSING = True
OVERWRITE = False

subjects_to_run = SUBJECTS
if ONLY_MISSING:
    subjects_to_run = [
        sid for sid in SUBJECTS
        if not get_epochs_path(sid, PIPELINE, 'feedback', root_dir=EPOCHS_ROOT).exists()
    ]

results = []
for subject_id in subjects_to_run:
    raw = _load_bids_raw(subject_id, Path(config.BIDS_ROOT[USER]), pipeline_name=PIPELINE)
    epochs, save_path, _ = build_and_save_feedback_epochs(
        raw=raw,
        subject_id=subject_id,
        pipeline_name=PIPELINE,
        bids_root=Path(config.BIDS_ROOT[USER]),
        root_dir=EPOCHS_ROOT,
        overwrite=OVERWRITE,
    )
    results.append({
        'subject_id': subject_id,
        'status': 'built',
        'n_epochs': len(epochs),
        'save_path': save_path,
    })

summary_df = pd.DataFrame(results, columns=['subject_id', 'status', 'n_epochs', 'save_path'])
summary_df
```
